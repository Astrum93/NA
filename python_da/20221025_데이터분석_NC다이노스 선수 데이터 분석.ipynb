#### 최종 데이터

2013, 2014, 2015 데이터를 모두 합치되 각각이 몇 년도 데이터인지 알 수 있도록 '연도' 값을 추가

```python
# nc13, nc14, nc15 변수에 '연도' 컬럼 추가
# nc13에는 2013을, nc14에는 2014를 스칼라 값으로..
# 이와 같이 연도 식별이 가능하게 컬럼을 추가해주세요.
```

```python
# 13년도

nc13['연도'] = 2013
nc13.head()
```

```python
nc14['연도'] = 2014
```

```python
nc15['연도'] = 2015
```

```python
# 여러개의 df 결합
# 결합 방식 : 컬럼개수 및 속성이 완전히 같으므로 row길이를 늘려준다.
# 최종 결과는 : 20 + 20 + 20 = (60,20)
```

```python
nc_result1 = pd.concat([nc13,nc14,nc15])
```

```python
nc_result1
```

```python
# 조건색인을 이용해 2013,2014 데이터만 조회해 보세요.
# nc_result1[(nc_result1['연도']==2013) / (nc_result1['연도']==2014)]
nc_result1[nc_result1['연도']!= 2015]
```

## 3. 2013~2015년 선수별 3년 평균 안타 수

- 평균 안타수가 높은 순서로 정렬해서 보이기

```python
hit_result = nc_result1.groupby('선수명')[['안타']].mean().sort_values(by='안타',ascending=False)
hit_result
```

```python
hit_result.head(10)
```

```python
hit_player = hit_result.head(10).index.tolist()
```

## 4. 평균 안타수 기준 상위 10위 타자들의 3년간 안타, 홈런, 타율 조회
#### 참고) NaN 값이 있을 경우 '-'로 대체해서 표현

```python
nc_result1.groupby('선수명')[['안타']].mean().sort_values(by='안타',ascending=False)
```

```python
# 조건색인시 .isin(리스트) 을 사용하면, 해당 리스트에 포함된 사람만 조회 가능합니다.
# fancy 인덱싱(조건색인에서 미리 준비된 리스트에 해당하는 목록만 남김)
nc_result1[nc_result1['선수명'].isin(hit_player)]
```

```python
top10_player = nc_result1[nc_result1['선수명'].isin(hit_player)]
top10_player
```

```python
# 연도, 선수명, 안타, 홈런, 타율 컬럼만 남겨주세요.
top10_player = top10_player[['연도','선수명','안타','홈런','타율']]
```

```python
top10_player
```

```python
# 연도, 선수명을 로우로 내리기
# df.set_index(['내릴 컬럼들...'])
top10_player = top10_player.set_index(['연도','선수명'])
top10_player
```

```python
# 위 자료에 대해 unstack()으로 "연도" 를 컬럼으로 올리겠습니다.
# NaN를 다른 델이터로 교체하기 : .fillna("채울자료")
top10_player.unstack(0).fillna("-")
```

```python
# 6.연도별 팀의 총 안타수, 총 홈런수, 총 도루수
# 구해서 보여주세요.
team_result=nc_result1.groupby('연도')[['안타','홈런','도루']].sum()
```

```python
# 엑셀이나 csv로 빼는 방법
# 엑셀 : .to_excel('경로 및 파일명')
# csv : .to_csv('경로 및 파일명')
team_result.to_excel("NC result.xlsx")
```
