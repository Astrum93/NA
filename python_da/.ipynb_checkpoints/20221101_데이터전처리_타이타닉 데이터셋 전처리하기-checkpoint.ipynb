##### map() / apply()

- 데이터프레임/시리즈 데이터타입에 대해서 함수 적용(동일한 작업을 반복)
    - 각 컬럼
    - 각 로우

```python
# 남자 : male, 여자 : female
# 남자 : 0, 여자 : 1
# 문자타입 범주형 데이터를 숫자타입 범주형 데이터로 변환
data.head()
```

```python
# 직접 컬럼도 추가하고, 값도 매칭시켜주시면 됩니다.
# 방법 1 : {'male':0,....}
# 방법 2 : 함수를 선언해 각 로우 값이 뭐냐에 따라 리턴을 0이나 1로
data['sex_num'] = -1
###### 방법 1
#data['sex_num'] = data['Sex'].map({male:0, 'female':1})
###### 방법 2 
data['sex_num'] = data['Sex'].map(change_sex)
```

```python
data.head()
```

```python
def change_sex(value):
    if value == 'male':
        return  0
    elif value == 'female':
        return  1
```

```python
data['sex_num']=data['Sex'].apply(change_sex)
```

```python
data.head()
```

### 3. 이름 길이 값을 저장하는 name_length 컬럼 추가

```python
def name_len(value):
    return len(value)
```

```python
data['name_len'] = -1
```

```python
data.head()
```

```python
data['name_len'] = data['Name'].map(name_len)
```

```python
data.head()
```

```python
# 방법 2
data['name_len'] = data['Name'].map(len)
```

- 파이썬 재장함수 : round()
- numpy 함수 : np.ceil()

```python
data['fare_ceil']= -1
data['fare_ceil']= data['Fare'].map(round)
data.head()
```

```python
# 운임요금이 생존여부에 영향을 줬는지 체크해주시고 근거를 같이 제출해주세요.

data[['Survived','fare_ceil']].corr()
```

```python
# np.ceil() : 소수점 첫 자리에 있는 모든 값(사사오입 안함)에 대해서 올림을 적용
# 특정값보다 큰 정수 중 최소 값으로 적용
data['Fare'].map(np.ceil)
```
