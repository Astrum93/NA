### 알라딘 크롤링


```python
# 필수 라이브러리를 임포트해주세요.


from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time
from selenium.webdriver.common.by import By
```

```python
# 전체 소스가 있고
# 태그가 div면서 클래스가 ss_book_box 이면 책 한권씩 포장된거


# 책 한권에서 이름 저자 가격만 남기려면
# 책한권의 div에서 li로 솎아주고
# li리스트에서 1,2,3번째가 이름 저자 가격이다.

# 그럼 알라딘 베스트 셀러 1페이지의 책들의 이름 저자 가격만 남기려면

# 1. 위 주소에 접근해서 소스를 저장하고
driver = webdriver.Chrome('chromedriver')
driver.get('https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&start=we')
time.sleep(2)
source = driver.page_source
driver.close()
```

```python
# 2. 소스를 파싱하고
parsed_source = BeautifulSoup(source, "html.parser")

# 3. 태그가 div면서 클래스가 ss_book_box로 솎아서 책들만 남기고
### div_books_list = 50개의 책들의 정보###
div_books_list= parsed_source.find_all("div", class_="ss_book_box")
```

```python
# div_books_list는 50개의 책들의 정보이다.
len(div_books_list)
```

```python
# 파싱된 소스에서 li태그가 몇갠지 조사해 주세요
len(parsed_source.find_all("li"))
```

```python
# bs4.resultest 자료형 find_all을 받을 수 없습니다.(리스트로 취급되기 때문)
# 따라서, 내부 요소 하나하나를 개별적으로 처리해야 합니다.
# 1. 인덱싱을 한다.
# 2. 반복문 내부에서 하나하나 돌아가면서 대입해서 처리한다.
li_books_list= div_books_list[0].find_all("li") # 인덱스0 = 1위책
```

```python
# 책들 전체를 div에서 li로 솎아주고 (div책들이 li 책들로 변함)
# li_books_list= div_books_list[0].find_all("li") # 인덱스0 = 1위책
for books in div_books_list:
    li_books_list = books.find_all("li")
    
    print(li_books_list[1].text)
    print(li_books_list[2].text)
    print(li_books_list[3].text)
    print("--------------------------")

```

```python
# li_books_list [1],[2],[3] 이 한권이다.
print(li_book_list[1].text)
print(li_book_list[2].text)
print(li_book_list[3].text)

book = []
for a in range(1,4):
    book.append(li_book_list[a].text.replace("\xa0", ""))
```

### 베스트 셀러 2페이지

- li_books2_list는 순서가 다르다 어떻게 뽑을까?

```python
driver = webdriver.Chrome('chromedriver')
driver.get('https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=2&cnt=1000&SortOrder=1')
time.sleep(2)
source = driver.page_source
driver.close()
```

```python
# 2. 소스를 파싱하고
parsed_source = BeautifulSoup(source, "html.parser")

# 3. 태그가 div면서 클래스가 ss_book_box로 솎아서 책들만 남기고
### div_books_list = 50개의 책들의 정보###
div_books_list= parsed_source.find_all("div", class_="ss_book_box")
```

```python
for book in div_books_list:
    li_books_list = book.find_all("li")
```

```python
# 2페이지 <li>에 증정품 정보가 없는 책은 어떻게 처리할지 생각해보세요.
if(li_books_list[1].text[0] == '['):
    print(li_books_list[1].text)
    print(li_books_list[2].text)
    print(li_books_list[3].text)
else:
    print(li_books_list[0].text)
    print(li_books_list[1].text)
    print(li_books_list[2].text)
print("------------------------------------------------------")
```
