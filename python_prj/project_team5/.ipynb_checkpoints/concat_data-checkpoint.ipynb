```python
import pandas as pd
df = pd.DataFrame()

for files in folders:
    df_b= pd.read_csv(files,encoding='cp949')
    df = pd.concat([df, df_b],axis=1)
df.head()
```

```python
del df['Unnamed: 0']
df.head()
```

```python
df=df[['datetime','가계예금총금액','주택매매가격지수','취업자수','가계대출금리','주택전세가격지수']]
df.head()
```
