# 가계 대출 금리는 Open API로 데이터 소싱 실패
# Open API 데이터소싱 가능한 한국은행 금리 연습

```python
# 사이트에 자료 요청
import urllib.request

# json 데이터 핸들링
import json

# DataFrame 자료형 활용
import pandas as pd

# json 데이터를 pandas DataFrame으로 변환
from pandas.io.json import json_normalize
```

#### 한국은행 금리 정보(data.go.kr)

### 1. 포메팅을 위해 변수 저장 및 URL 저장
### 2. 포메팅 변수는 년도의 시작 및 끝 's' 와 'f' 로 저장.

```python
key = "A3BU973BS19JTFZNBEQX"
s = '200101'
f = '202210'
url = "https://ecos.bok.or.kr/api/StatisticSearch/%s/json/kr/1/264/722Y001/M/%s/%s/0101000/?/?/?" % (key, s, f)
url
```

### 3. result 변수에 데이터 request 메서드 저장.

```python
result = urllib.request.urlopen(url)
```

```python
json_raw_data = result.read()
```

```python
result.read()
```

### 4. 소싱한 데이터 확인

```python
json_raw_data
```

### 5. 소싱한 데이터 타입 확인

```python
type(json_raw_data)
```

### 6. 소싱한 데이터 'utf-8-sig'형식으로 데이터처리 했다.

```python
# utf-8-sig 
json_utf8 = json_raw_data.decode('utf-8-sig')
```

### 확인

```python
json_utf8
```

### 타입 확인

```python
type(json_utf8)
```

### 7. json.loads를 이용해 파이썬 객체로 역 직렬화

```python
json_complete = json.loads(json_utf8)
```

```python
json_complete
```

### 8. 구조확인을 해보니 key값이 3중 dict데이터 타입이었다.
### 따라서 인덱싱으로 원하는 자료에 접근

```python
json_com = json_complete['StatisticSearch'].copy()
```

```python
json_com
```

### 9. key값을 'row'로 하는 value값이 내가 원하던 자료인것을 확인

```python
a= json_com['row'].copy()
```

```python
type(a)
```

### 10. 원하는 dict자료를 final변수에 DataFrame화 하고 .copy()까지 써서 깊은복사완료.

```python
final = pd.DataFrame(a).copy()
```

### 11. final 자료 확인

```python
final
```

### 데이터 타입까지 확인

```python
type(final)
```
