```python
#     
from bs4 import BeautifulSoup   # 
from selenium import webdriver  #   
import requests #requests   API 
import time             #    time 


from selenium.webdriver.common.by import By
import codecs
```

```python
driver = webdriver.Chrome()
time.sleep(1)
driver.get("https://ecos.bok.or.kr/#/Short/22a8df") #  url 
time.sleep(2)
```

$4.$ **     **<br> 
-  1 : result_dict = {} date_listd value_list      <br>
-  8 : date_list = list() rgheader    <br>
-  17 : value_list = list() rglist    

```python
result_dict = {}  # [date_list[i]] = value_list[i] 
source = driver.page_source
parsed_source = BeautifulSoup(source, "html.parser")

# 
rgheader = parsed_source.find_all("div", class_="rg-header-renderer")

date_list = list() # date_list :  

for i in range(262):            # 2000 1  2022 10    262    range(262) 
    date_list.append(rgheader[i].text.replace('/','')) # .replace('/','')      2000/01 200001  date_list i   


# 
rglist = parsed_source.find_all("div", class_="rg-renderer")

value_list = list() # value_list:  

for i in range(262):         # 2000 1  2022 10    262    range(262) 
    value_list.append(rglist[i].text) # date_list i = (262)   append value_list 
    result_dict[date_list[i]] = value_list[i] # date_list i = (262)   append value_list 


tmp_dict = result_dict.copy()  # tmp_dict   
```
