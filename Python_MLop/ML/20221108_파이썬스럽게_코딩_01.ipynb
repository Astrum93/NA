# 리스트 컴프리헨션

```python
c_list = ['apple', 'banana', 'cherry', 'pear', 'bear']
```

```python
new_list = []
for c in c_list:
    new_list.append(c[0])
new_list
```

```python
[c for c in c_list]
```

```python
[len(c) for c in c_list]
```

```python
def welcome(v):
    print(f"환영합니다! {v}")

names = ["강동원", "김연아", "페이커"]
[welcome(n) for n in names]
```

```python
#@title 삼항연산자
"O" if True else "X"
```

```python
boolean = False
"O" if boolean else "X"
```

```python
#@title 리스트 컴프리헨션에서의 조건문
[c for c in c_list if len(c) > 5]
```

```python
#@title 리스트 컴프리헨션에서의 삼항연산자 (주의)
[c if len(c) > 5 else "Sorry" for c in c_list]
```
